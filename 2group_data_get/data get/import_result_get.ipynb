```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.dates as mdates

import math
```

```python
#加载时间
timelist = []
for i in range(12):
    timelist.append(str(201101 + i))
    
for i in range(12):
    timelist.append(str(201201 + i))
    
for i in range(12):
    timelist.append(str(201301 + i))

for i in range(12):
    timelist.append(str(201401 + i))
    
for i in range(12):
    timelist.append(str(201501 + i))
    
for i in range(12):
    timelist.append(str(201601 + i))

for i in range(12):
    timelist.append(str(201701 + i))
    
for i in range(12):
    timelist.append(str(201801 + i))

for i in range(12):
    timelist.append(str(201901 + i))
    
for i in range(12):
    timelist.append(str(202001 + i))
    
for i in range(5):
    timelist.append(str(202101 + i))
    
```

```python
#将所有日期数据合并
alldata = pd.DataFrame(columns=['rtTitle', 'ptTitle', 'TradeValue', 'monthdate'])
for i in timelist:
    result = pd.read_csv(i + '_APEC_import_trade.csv')
    result['monthdate'] = i
    alldata=pd.concat([alldata, result])
    
#获取每个国家每个月进口总额
alldata['TradeValue'] = pd.to_numeric(alldata['TradeValue'], downcast='float')
month_total = alldata[alldata['ptTitle'] == 'World']
month_total['TradeValue'] = month_total['TradeValue']/(10 ** 9)
#获取国家名code映射
counrty_code = pd.read_csv('nodes.csv')
code_book = pd.read_csv('codebook.csv', header=None, sep="	")
code_book.columns = ['0','country_code','Label','3','4','5','6','7']

#获取各个国家的code
month_total.columns=['Id', 'ptTitle', 'TradeValue', 'monthdate']
month_total_2 = pd.merge(counrty_code, month_total, on='Id',how='right')
month_total_id = pd.merge(code_book, month_total_2, on='Label',how='inner')
month_total_id['monthdate'] = pd.to_datetime(month_total_id['monthdate'].astype('str'),format = '%Y%m')
```

```python
#保存结果
new_dataframe = month_total_id['country_code'].unique()
new_dataframe.sort()
result_frame = ['month']
for i in new_dataframe:
    #去除塞尔维亚和黑山 SC
    if i == 'SC':
        continue
    result_frame.append(i)
result_pd = pd.DataFrame(columns = result_frame)
```

```python
#将国家转化为列输出，空值为0
import_data = month_total_id[['monthdate','country_code','TradeValue']]
for month in import_data['monthdate'].unique():
    fill_country_list = [month]
    for country in new_dataframe:
        #去除塞尔维亚和黑山 SC
        if country == 'SC':
            continue
        if(import_data[(import_data['monthdate'] == month) & (import_data['country_code'] == country)].empty == 0):
            fill_country_list.append(import_data[(import_data['monthdate'] == month) & (import_data['country_code'] == country)].iloc[0,2])
        else:
            fill_country_list.append(0)
    result_pd.loc[len(result_pd)] = fill_country_list
```

```python
pd.set_option('display.max_rows', 20)
pd.set_option('display.max_columns', 20)
result_pd.to_csv('201101_202105_import.csv',index = False)
```
