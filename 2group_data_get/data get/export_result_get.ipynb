```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.dates as mdates

import math
```

```python
#加载时间
timelist = []
for i in range(12):
    timelist.append(str(201101 + i))
    
for i in range(12):
    timelist.append(str(201201 + i))
    
for i in range(12):
    timelist.append(str(201301 + i))

for i in range(12):
    timelist.append(str(201401 + i))
    
for i in range(12):
    timelist.append(str(201501 + i))
    
for i in range(12):
    timelist.append(str(201601 + i))

for i in range(12):
    timelist.append(str(201701 + i))
    
for i in range(12):
    timelist.append(str(201801 + i))

for i in range(12):
    timelist.append(str(201901 + i))
    
for i in range(12):
    timelist.append(str(202001 + i))
    
for i in range(5):
    timelist.append(str(202101 + i))
    
```

```python
#将所有日期数据合并
ex_alldata = pd.DataFrame(columns=['rtTitle', 'ptTitle', 'TradeValue', 'monthdate'])
for i in timelist:
    ex_result = pd.read_csv(i + '_APEC_export_trade.csv')
    ex_result['monthdate'] = i
    ex_alldata=pd.concat([ex_alldata, ex_result])
    
#获取每个国家每个月出口总额
ex_alldata['TradeValue'] = pd.to_numeric(ex_alldata['TradeValue'], downcast='float')
ex_month_total = ex_alldata[ex_alldata['ptTitle'] == 'World']
ex_month_total['TradeValue'] = ex_month_total['TradeValue']/(10 ** 9)
#获取国家名code映射
counrty_code = pd.read_csv('nodes.csv')
code_book = pd.read_csv('codebook.csv', header=None, sep="	")
code_book.columns = ['0','country_code','Label','3','4','5','6','7']

#获取各个国家的code
ex_month_total.columns=['Id', 'ptTitle', 'TradeValue', 'monthdate']
ex_month_total_2 = pd.merge(counrty_code, ex_month_total, on='Id',how='right')
ex_month_total_id = pd.merge(code_book, ex_month_total_2, on='Label',how='inner')
ex_month_total_id['monthdate'] = pd.to_datetime(ex_month_total_id['monthdate'].astype('str'),format = '%Y%m')
```

```python
#保存结果
new_dataframe = ex_month_total_id['country_code'].unique()
new_dataframe.sort()
result_frame = ['month']
for i in new_dataframe:
    #去除塞尔维亚和黑山 SC
    if i == 'SC':
        continue
    result_frame.append(i)
result_pd = pd.DataFrame(columns = result_frame)
#将国家转化为列输出，空值为0
export_data = ex_month_total_id[['monthdate','country_code','TradeValue']]
for month in export_data['monthdate'].unique():
    fill_country_list = [month]
    for country in new_dataframe:
        #去除塞尔维亚和黑山 SC
        if country == 'SC':
            continue
        if(export_data[(export_data['monthdate'] == month) & (export_data['country_code'] == country)].empty == 0):
            fill_country_list.append(export_data[(export_data['monthdate'] == month) & (export_data['country_code'] == country)].iloc[0,2])
        else:
            fill_country_list.append(0)
    result_pd.loc[len(result_pd)] = fill_country_list
```

```python
result_pd.to_csv('201101_202105_export.csv',index = False)
```
