```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.dates as mdates

import math
```

```python
#加载时间
timelist = []
for i in range(11):
    timelist.append(str(201502 + i))
    
for i in range(12):
    timelist.append(str(201601 + i))

for i in range(12):
    timelist.append(str(201701 + i))
    
for i in range(12):
    timelist.append(str(201801 + i))

for i in range(12):
    timelist.append(str(201901 + i))
    
for i in range(12):
    timelist.append(str(202001 + i))
    
for i in range(5):
    timelist.append(str(202101 + i))
    
    
#将所有日期数据合并
alldata = pd.DataFrame(columns=['rtTitle', 'ptTitle', 'TradeValue', 'monthdate'])
for i in timelist:
    result = pd.read_csv(i + '_APEC_import_trade.csv')
    result['monthdate'] = i
    alldata=pd.concat([alldata, result])

#获取ex_data作为补充
ex_alldata = pd.DataFrame(columns=['rtTitle', 'ptTitle', 'TradeValue', 'monthdate'])
for i in timelist:
    result = pd.read_csv(i + '_APEC_export_trade.csv')
    result['monthdate'] = i
    ex_alldata=pd.concat([ex_alldata, result])


#读取统一名称
APEC_country = pd.read_csv('APEC_country.csv')
APEC_country.iloc[1,4] = 'China, Hong Kong SAR'
APEC_country.iloc[9,4] = 'Rep. of Korea'
APEC_country = APEC_country[['un_name','en_name']]
APEC_country.columns=['rtTitle','en_name']

result_name = 'TradeValue'

#国家名映射
result = pd.merge(APEC_country, alldata, on='rtTitle',how='inner')
result = result[['en_name','monthdate', 'ptTitle', result_name]]
result.columns=['country1', 'monthdate', 'rtTitle', result_name]
result = pd.merge(APEC_country, result, on='rtTitle',how='inner')
result = result[['monthdate','country1', 'en_name', result_name]]
result.columns=['monthdate','country1', 'country2', result_name]

#出口国家名映射
ex_result = pd.merge(APEC_country, ex_alldata, on='rtTitle',how='inner')
ex_result = ex_result[['en_name','monthdate', 'ptTitle', result_name]]
ex_result.columns=['country1', 'monthdate', 'rtTitle', result_name]
ex_result = pd.merge(APEC_country, ex_result, on='rtTitle',how='inner')
ex_result = ex_result[['monthdate','country1', 'en_name', result_name]]
ex_result.columns=['monthdate','country1', 'country2', result_name]
```

```python
#加载国家
country_list = APEC_country['en_name'].unique()
country_list.sort()

#表头
result_frame = ['time','country']
for i in country_list:
    result_frame.append(i)

    
result_pd = pd.DataFrame(columns = result_frame)
```

```python
country_list
```

```python
result_pd = pd.DataFrame(columns = result_frame)
alldata = result
for time in timelist:
    count = 0
    for country_A in country_list:
        now_list = [time]
        now_list.append(country_list[count])
        count += 1
        for country_B in country_list:
            if(country_A == country_B):
                now_list.append(0)
            elif(alldata[(alldata['country1'] == country_A) & (alldata['country2'] == country_B)& (alldata['monthdate'] == time)].empty == 0):
                now_list.append(alldata[(alldata['country1'] == country_A) & (alldata['country2'] == country_B)& (alldata['monthdate'] == time)].iloc[0,3])
            elif(ex_result[(ex_result['country1'] == country_B) & (ex_result['country2'] == country_A)& (ex_result['monthdate'] == time)].empty == 0):
                now_list.append(ex_result[(ex_result['country1'] == country_B) & (ex_result['country2'] == country_A)& (ex_result['monthdate'] == time)].iloc[0,3])
            else:
                now_list.append(-1)
        result_pd.loc[len(result_pd)] = now_list
    print(time)
    #break
```

```python
result_pd.to_csv('201502_202105_trade.csv',index = False)
```
