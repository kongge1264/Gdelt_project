```python
import pandas as pd
#插入越南读取19个国家
wto_export = pd.read_csv('2011_2021_wto_export.csv')
wto_import = pd.read_csv('2011_2021_wto_import.csv')
wto_export_data = wto_export[['Year','Period','Reporting Economy','Value']]
wto_import_data = wto_import[['Year','Period','Reporting Economy','Value']]

wto_export_data['Value'] = wto_import_data['Value']/(10 ** 3)
wto_import_data['Value'] = wto_import_data['Value']/(10 ** 3)
#import需要添加朝鲜
vm_import = pd.read_csv('vmimport.csv')
vm_import = vm_import[['Year','Period','Reporting Economy','Value']]
wto_import_data = wto_import_data.append(vm_import)
#分解数据
month_list = wto_export_data['Period'].unique()
country_list = wto_export_data['Reporting Economy'].unique()
country_list.sort()
year_list = wto_export_data['Year'].unique()
year_list.sort()
```

```python
APEC_country = pd.read_csv('APEC_country.csv')
```

```python
#表头
result_frame = ['month']
for i in country_list:
    result_frame.append(i)
    
im_result_pd = pd.DataFrame(columns = result_frame)
ex_result_pd = pd.DataFrame(columns = result_frame)
#进出口一起插入
for year in year_list:
    for month in range(12):
        if(month < 9):
            im_fill_country_list = [str(year) + '0' + str(month+1)]
            ex_fill_country_list = [str(year) + '0' + str(month+1)]
        else:
            im_fill_country_list = [str(year) + str(month+1)]
            ex_fill_country_list = [str(year) + str(month+1)]
        for country in country_list:
            im_fill_country_list.append(wto_import_data[(wto_import_data['Year'] == year) &(wto_import_data['Period'] == month_list[month]) & (wto_import_data['Reporting Economy'] == country)].iloc[0,3])
            ex_fill_country_list.append(wto_export_data[(wto_export_data['Year'] == year) &(wto_export_data['Period'] == month_list[month]) & (wto_export_data['Reporting Economy'] == country)].iloc[0,3])
        im_result_pd.loc[len(im_result_pd)] = im_fill_country_list
        ex_result_pd.loc[len(ex_result_pd)] = ex_fill_country_list
```

```python
im_result_pd.to_csv('201101_202105_import_wto.csv',index = False)
ex_result_pd.to_csv('201101_202105_export_wto.csv',index = False)
```
