```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.dates as mdates

import math
```

```python
file_name = '201502_202105_APEC_reportnums.csv'
```

```python
file_name = '201502_202105_APEC_2reportnums.csv'
```

```python
#读取数据
alldata = pd.read_csv(file_name)
alldata['monthdate'] = pd.to_numeric(alldata['monthdate'], downcast='integer')
country_list = alldata['en_name'].unique()
country_list.sort()

timelist = alldata['monthdate'].unique()
timelist.sort()
#表头
result_frame = ['time','country']
for i in country_list:
    result_frame.append(i)

    
result_pd = pd.DataFrame(columns = result_frame)
```

```python
#国家报道另一个国家运算
result_pd = pd.DataFrame(columns = result_frame)
for time in timelist:
    for country_A in country_list:
        now_list = [time]
        now_list.append(country_A)
        for country_B in country_list:
            if(alldata[(alldata['en_name'] == country_A) & (alldata['en_name_1'] == country_B)& (alldata['monthdate'] == time)].empty == 0):
                now_list.append(alldata[(alldata['en_name'] == country_A) & (alldata['en_name_1'] == country_B)& (alldata['monthdate'] == time)].iloc[0,3])
            else:
                now_list.append(0)
        result_pd.loc[len(result_pd)] = now_list
```

```python
#两个国家共同提及运算，需要查询对角位置
result_pd = pd.DataFrame(columns = result_frame)
for time in timelist:
    for country_A in country_list:
        now_list = [time]
        now_list.append(country_A)
        for country_B in country_list:
            if(alldata[(alldata['en_name'] == country_A) & (alldata['en_name_1'] == country_B)& (alldata['monthdate'] == time)].empty == 0):
                now_list.append(alldata[(alldata['en_name'] == country_A) & (alldata['en_name_1'] == country_B)& (alldata['monthdate'] == time)].iloc[0,3])
            #在搜索一次，获取对角关系
            elif(alldata[(alldata['en_name'] == country_B) & (alldata['en_name_1'] == country_A)& (alldata['monthdate'] == time)].empty == 0):
                now_list.append(alldata[(alldata['en_name'] == country_B) & (alldata['en_name_1'] == country_A)& (alldata['monthdate'] == time)].iloc[0,3])
            else:
                now_list.append(0)
        result_pd.loc[len(result_pd)] = now_list
        
result_pd.to_csv('201502_202105_2mentions.csv',index = False)
```
