```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.dates as mdates

import math
```

```python
#加载时间
time = [2015,2017,2019,2021]


#将所有日期数据合并
alldata = pd.DataFrame(columns=['monthdate', 'country_code', 'report_code', 'report_portation'])
for i in range(3):
    result = pd.read_csv(str(time[i])+ '_'+str(time[i+1])+'.csv')  
    alldata=pd.concat([alldata, result])

APEC_country = pd.read_csv('APEC_country.csv')
APEC_country.iloc[1,4] = 'China, Hong Kong SAR'
APEC_country = APEC_country[['country_code','en_name']]

#国家名映射
result = pd.merge(APEC_country, alldata, on='country_code',how='inner')
result = result[['en_name','monthdate', 'report_code', 'report_portation']]
result.columns=['country1', 'monthdate', 'country_code', 'report_portation']
result = pd.merge(APEC_country, result, on='country_code',how='inner')
result = result[['monthdate','country1', 'en_name', 'report_portation']]
result.columns=['monthdate','country1', 'country2', 'report_portation']
```

```python
#去掉文莱
country_list = APEC_country['en_name'].unique()
country_list.sort()
country_list = np.delete(country_list, 1)

#表头
result_frame = ['time']
for i in country_list:
    for j in country_list:
        if(i != j):
            result_frame.append(i + '*' + j)

timelist = result['monthdate'].unique()
timelist.sort()
result_pd = pd.DataFrame(columns = result_frame)
```

```python
len(result_pd.columns)
```

```python
result_pd = pd.DataFrame(columns = result_frame)
for time in timelist:
    now_list = [time]
    for country_A in country_list:
        for country_B in country_list:
            if(country_A == country_B):
                continue
            elif(result[(result['country1'] == country_A) & (result['country2'] == country_B)& (result['monthdate'] == time)].empty == 0):
                now_list.append(result[(result['country1'] == country_A) & (result['country2'] == country_B)& (result['monthdate'] == time)].iloc[0,3])
            else:
                now_list.append(-1)
    #print(len(now_list))
    result_pd.loc[len(result_pd)] = now_list
    print(time)
    #break
```

```python
result_pd.to_csv('news_relations.csv',index = False)
```
