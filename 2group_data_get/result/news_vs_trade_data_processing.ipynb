## Full year

```python
df_trade_fy = df_trade.iloc[:,2:]
df_mention_fy = df_mention.iloc[:,2:]
df_2mentions_fy = df_2mentions.iloc[:,2:]

print()

assert df_trade_fy.shape == df_mention_fy.shape == df_2mentions_fy.shape

Y = []
X = []

for i in range(df_trade_fy.shape[0]):
    for j in range(df_trade_fy.shape[1]):
        if df_trade_fy.iloc[i,j] > 0 and df_mention_fy.iloc[i,j] > 0 and df_2mentions_fy.iloc[i,j] > 0:
            X.append([df_mention_fy.iloc[i,j],df_2mentions_fy.iloc[i,j]])
            Y.append(df_trade_fy.iloc[i,j])
```

```python
import numpy as np
X = np.array(X)
print(X.shape)
Y = np.array(Y)
print(Y.shape)
```

```python
Y.reshape(-1,1).shape
```

```python
print('mention vs trade:')
print(np.corrcoef(X[:,0],Y))
print('R square =', np.corrcoef(X[:,0],Y)[0,1]**2)
```

```python
print('2mentions vs trade:')
print(np.corrcoef(X[:,1],Y))
print('R square =', np.corrcoef(X[:,1],Y)[0,1]**2)
```
